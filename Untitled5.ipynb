```python
import pandas as pd
```

```python
btc = pd.read_csv('crypto_prices.csv',parse_dates=['Date'])
eth = pd.read_csv('crypto_prices_ETH.csv',parse_dates=['Date'])
xrp = pd.read_csv('crypto_prices_XRP.csv',parse_dates=['Date'])
sp500 = pd.read_csv('sp500.csv',parse_dates=['Дата'])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.plot(btc['Date'], btc['Close'],
         color='#1f77b4',
         linewidth=1.5)
plt.title('Динамика цены BTC', pad=20, fontsize=14)
plt.xlabel('Дата', labelpad=10)
plt.ylabel('Цена (USD)', labelpad=10)
plt.grid(alpha=0.3, linestyle='--')
plt.gcf().autofmt_xdate()
plt.savefig('price_chart.png', dpi=200, bbox_inches='tight')
plt.show()
```
