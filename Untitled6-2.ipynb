```python
# Блок 1. Монтаж Google Drive и загрузка данных
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr, shapiro
from statsmodels.tsa.stattools import grangercausalitytests, adfuller, coint
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
import scipy.stats as stats

# Загрузка данных (файлы содержат столбцы: Date, Close)
sp500 = pd.read_csv('/content/drive/MyDrive/data/sp500.csv', parse_dates=['Date'])
btc   = pd.read_csv('/content/drive/MyDrive/data/crypto_prices.csv', parse_dates=['Date'])
eth   = pd.read_csv('/content/drive/MyDrive/data/crypto_prices_ETH.csv', parse_dates=['Date'])
xrp   = pd.read_csv('/content/drive/MyDrive/data/crypto_prices_XRP.csv', parse_dates=['Date'])

# Переименование столбца Close в соответствующее имя актива
sp500.rename(columns={'Close': 'SP500'}, inplace=True)
btc.rename(columns={'Close': 'BTC'}, inplace=True)
eth.rename(columns={'Close': 'ETH'}, inplace=True)
xrp.rename(columns={'Close': 'XRP'}, inplace=True)

# Объединение данных по столбцу Date
df = sp500.merge(btc[['Date', 'BTC']], on='Date', how='inner')
df = df.merge(eth[['Date', 'ETH']], on='Date', how='inner')
df = df.merge(xrp[['Date', 'XRP']], on='Date', how='inner')
df.sort_values('Date', inplace=True)
df.reset_index(drop=True, inplace=True)

# Расчет дневной доходности
for col in ['SP500', 'BTC', 'ETH', 'XRP']:
    df[col + '_ret'] = df[col].pct_change()
df.dropna(inplace=True)
```

```python
# Блок 2. Визуализация цен и нормализация

# График цен с логарифмическим масштабом
plt.figure(figsize=(10,6))
plt.plot(df['Date'], df['SP500'], label='SP500')
plt.plot(df['Date'], df['BTC'], label='BTC')
plt.plot(df['Date'], df['ETH'], label='ETH')
plt.plot(df['Date'], df['XRP'], label='XRP')
plt.yscale('log')
plt.xlabel('Date'); plt.ylabel('Цена (log)'); plt.title('Цены активов')
plt.legend(); plt.show()

# Нормализация цен (начало = 1)
df['SP500_norm'] = df['SP500'] / df['SP500'].iloc[0]
df['BTC_norm']   = df['BTC']   / df['BTC'].iloc[0]
df['ETH_norm']   = df['ETH']   / df['ETH'].iloc[0]
df['XRP_norm']   = df['XRP']   / df['XRP'].iloc[0]

plt.figure(figsize=(10,6))
plt.plot(df['Date'], df['SP500_norm'], label='SP500')
plt.plot(df['Date'], df['BTC_norm'], label='BTC')
plt.plot(df['Date'], df['ETH_norm'], label='ETH')
plt.plot(df['Date'], df['XRP_norm'], label='XRP')
plt.xlabel('Date'); plt.ylabel('Нормированная цена'); plt.title('Нормированные цены')
plt.legend(); plt.show()
```

```python
# Блок 3. Корреляционный анализ доходностей

# Матрица корреляций
corr = df[['SP500_ret', 'BTC_ret', 'ETH_ret', 'XRP_ret']].corr()
print("Матрица корреляций:\n", corr)

# Корреляция SP500_ret с криптовалютами
for asset in ['BTC_ret', 'ETH_ret', 'XRP_ret']:
    r, p = pearsonr(df['SP500_ret'], df[asset])
    print(f"SP500_ret vs {asset}: r = {r:.3f}, p = {p:.3f}")
```

SP500_ret слабо коррелирует с криптовалютами (r ~ 0.08–0.12).

Между криптоактивами связь выше, особенно BTC и ETH (r ~ 0.80).

Корреляции статистически значимы (p < 0.05).

Фондовый рынок и криптовалюты движутся в основном независимо, но крипта внутри себя более связана.

```python
# Блок 4. Rolling корреляция (динамика связи SP500 и BTC)
window = 30
df['rolling_corr_SP500_BTC'] = df['SP500_ret'].rolling(window=window).corr(df['BTC_ret'])
plt.figure(figsize=(10,6))
plt.plot(df['Date'], df['rolling_corr_SP500_BTC'])
plt.xlabel('Date'); plt.ylabel('Rolling Corr')
plt.title(f'Rolling корреляция SP500 и BTC (окно = {window} дней)')
plt.show()
```

Корреляция заметно колеблется от около 0 до 0.4–0.5, иногда уходя в отрицательную зону.

Связь между SP500 и BTC не стабильна: в одни периоды они движутся более синхронно, в другие — почти независимо.

Макроэкономические события могут кратковременно усиливать или ослаблять корреляцию.







```python
# Блок 5. Тест Грейнджера (проверка причинности: SP500_ret -> BTC_ret)
# Вызываем тест Грейнджера без вывода в консоль (verbose=False)
results = grangercausalitytests(df[['BTC_ret','SP500_ret']], maxlag=5, verbose=False)

# Извлекаем p-значения для SSR-based F test по каждому лагу
p_vals = []
for lag, val in results.items():
    p_val = val[0]['ssr_ftest'][1]  # p-value из SSR-based F test
    p_vals.append((lag, p_val))

# Сортируем по лагу (хотя обычно и так по порядку)
p_vals.sort(key=lambda x: x[0])

lags = [item[0] for item in p_vals]
values = [item[1] for item in p_vals]

# Строим график
plt.figure(figsize=(6,4))
plt.bar(lags, values, color='skyblue', edgecolor='black')
plt.axhline(0.05, color='red', linestyle='--', label='Уровень значимости = 0.05')
plt.xlabel('Lag')
plt.ylabel('p-value')
plt.title('Granger Causality p-values (SP500_ret -> BTC_ret)')
plt.legend()
plt.show()
```

Высокие p-значения при всех лагах (0.2–0.97) говорят о том, что нет статистически значимого доказательства влияния прошлых значений SP500_ret на будущие значения BTC_ret.

Иными словами, не выявлена грейнджер-причинность от SP500 к биткойну в рамках данного теста и выбранных лагов (1–5).

Это значит, что история доходности SP500 не даёт дополнительной информации для прогноза доходности BTC.

```python
# Блок 6. Тест коинтеграции (Engle–Granger) для цен SP500 и BTC
score, p_value, _ = coint(df['SP500'], df['BTC'])
print("Коинтеграция SP500 и BTC: p-value =", p_value)
```

p-value = 0.0388 (меньше 0.05) свидетельствует о наличии коинтеграции между SP500 и BTC. То есть, несмотря на краткосрочные колебания, существует устойчивое долгосрочное соотношение между их уровнями цен.

```python
# Блок 7. Множественная регрессия: прогноз BTC_ret по SP500_ret и ETH_ret
X_multi = df[['SP500_ret', 'ETH_ret']]
y_multi = df['BTC_ret']
model_multi = LinearRegression().fit(X_multi, y_multi)
print("Множественная регрессия BTC_ret ~ SP500_ret + ETH_ret")
print("Коэффициенты:", model_multi.coef_)
print("Свободный член:", model_multi.intercept_)
```

```python
# Предсказания
y_pred = model_multi.predict(X_multi)

# График Предсказанное vs. Фактическое
plt.figure(figsize=(6,4))
plt.scatter(y_pred, y_multi, alpha=0.5, edgecolor='k')
plt.xlabel('Предсказанные BTC_ret')
plt.ylabel('Фактические BTC_ret')
plt.title('Сравнение предсказанных и фактических доходностей BTC')
plt.show()

r2_score = model_multi.score(X_multi, y_multi)
print("R^2 модели:", r2_score)

```

Уровень объяснения: Модель с двумя факторами (SP500_ret и ETH_ret) объясняет ~65% вариации доходности BTC, что довольно много для финансовых данных.

Сильная роль ETH_ret: Коэффициент при ETH_ret выше и положительный, указывая на более тесную связь между доходностями BTC и ETH по сравнению со SP500.

Ограниченность модели: Остаётся ещё ~35% неучтённых факторов (новостной фон, макроэкономика, другие криптоактивы).

Нет доказательства причинности: Модель указывает на статистическую связь, а не на то, что ETH «идёт за» BTC или наоборот. Они могут реагировать на общие рыночные драйверы.

```python
# Блок 8. Анализ распределения доходностей BTC

# Гистограмма BTC_ret
plt.figure(figsize=(10,6))
plt.hist(df['BTC_ret'], bins=50, density=True)
plt.xlabel('BTC_ret'); plt.ylabel('Плотность');
plt.title('Гистограмма BTC_ret')
plt.show()

# QQ-плот
sm.qqplot(df['BTC_ret'], line='s')
plt.title('QQ-плот BTC_ret')
plt.show()

# Тест Шапиро–Уилка
stat, p = shapiro(df['BTC_ret'])
print("Shapiro-Wilk для BTC_ret: stat =", stat, ", p =", p)
```

Отрицание нормальности: Очень низкое p-value (порядка
10^28) по тесту Шапиро–Уилка говорит о том, что доходности BTC статистически значимо отличаются от нормального распределения.

Толстые хвосты: На QQ-плоте видно заметные отклонения в области хвостов, что указывает на повышенную волатильность и экстремальные значения (характерно для крипторынков).

Практическое значение: Модели, предполагающие нормальность (например, классический VAR), могут работать неточно, так как реальное распределение доходностей BTC имеет более «тяжёлые» хвосты.\

```python
# Блок 9. Анализ волатильности BTC

# 30-дневное скользящее стандартное отклонение (волатильность)
df['BTC_vol'] = df['BTC_ret'].rolling(window=30).std()
plt.figure(figsize=(10,6))
plt.plot(df['Date'], df['BTC_vol'])
plt.xlabel('Date'); plt.ylabel('Волатильность (30-дневная)')
plt.title('Волатильность BTC_ret')
plt.show()
```

Переменная природа риска: Волатильность BTC_ret меняется во времени (2–10%), подтверждая непостоянный уровень риска криптовалют.

Всплески в кризисы: Наиболее высокие значения волатильности совпадают с рыночными шоками, указывая на чувствительность BTC к внешним факторам.

Учет волатильности в анализе: При прогнозировании и формировании портфеля важно учитывать периоды повышенной нестабильности крипторынка.







```python
# Блок 10. Портфельный анализ: портфель SP500 + BTC (равные веса)
df['portfolio_ret'] = (df['SP500_ret'] + df['BTC_ret']) / 2
port_mean = df['portfolio_ret'].mean()
port_std = df['portfolio_ret'].std()
sharpe_ratio = port_mean / port_std * np.sqrt(252)  # 252 торговых дня
print("Портфель SP500+BTC:")
print("Средняя доходность =", port_mean)
print("Стандартное отклонение =", port_std)
print("Sharpe ratio =", sharpe_ratio)
```

```python
df['portfolio_cum_ret'] = (1 + df['portfolio_ret']).cumprod()
plt.figure(figsize=(8,5))
plt.plot(df['Date'], df['portfolio_cum_ret'], label='SP500+BTC')
plt.xlabel('Date')
plt.ylabel('Кумулятивная доходность')
plt.title('Кумулятивная доходность портфеля SP500+BTC')
plt.legend()
plt.show()
```

Средняя доходность ~ 0.13% в день при стандартном отклонении ~ 2.18%.

Sharpe ratio ~ 0.97 — умеренно высокий показатель, указывающий на неплохое соотношение «доходность/риск».

Практический смысл: совместное владение SP500 и BTC может повышать потенциальную доходность по сравнению с отдельными активами, но остаётся достаточно высокий уровень волатильности.

```python
# Блок 11. Сравнение трендов: скользящие средние (30-дневные) для SP500 и BTC с логарифмической шкалой
df['SP500_MA30'] = df['SP500'].rolling(window=30).mean()
df['BTC_MA30']   = df['BTC'].rolling(window=30).mean()

plt.figure(figsize=(10,6))
plt.plot(df['Date'], df['SP500'], label='SP500')
plt.plot(df['Date'], df['SP500_MA30'], label='SP500 MA30', linestyle='--')
plt.plot(df['Date'], df['BTC'], label='BTC')
plt.plot(df['Date'], df['BTC_MA30'], label='BTC MA30', linestyle='--')
plt.xlabel('Date')
plt.ylabel('Цена')
plt.title('Скользящие средние (30 дней) с логарифмической шкалой')
plt.yscale('log')
plt.legend()
plt.show()
```

Более высокая волатильность BTC: Логарифмический масштаб и скользящие средние показывают, что биткойн растёт быстрее, но и колеблется сильнее, чем SP500.

Сглаживание не скрывает тренда: 30-дневные MA сглаживают краткосрочные колебания, однако контраст в динамике и амплитуде движений всё равно заметен.

Оба актива в восходящем тренде: Несмотря на разную волатильность, долгосрочный тренд и SP500, и BTC остаётся положительным.

```python
# Блок 12. Проверка стационарности (ADF-тест)

# Для доходностей
adf_sp500_ret = adfuller(df['SP500_ret'])
print("ADF тест для SP500_ret: p-value =", adf_sp500_ret[1])
adf_btc_ret = adfuller(df['BTC_ret'])
print("ADF тест для BTC_ret: p-value =", adf_btc_ret[1])

# Для цен
adf_sp500 = adfuller(df['SP500'])
print("ADF тест для SP500 цены: p-value =", adf_sp500[1])
adf_btc = adfuller(df['BTC'])
print("ADF тест для BTC цены: p-value =", adf_btc[1])
```

Доходности стационарны: Очень низкие p-value (почти 0) для SP500_ret и BTC_ret указывают на то, что ряды доходностей стационарны. Это типично для финансовых временных рядов (разность цен обычно стационарна).

Цены нестационарны: Высокие p-value (0.92 и 0.64) для уровней цен SP500 и BTC говорят о том, что эти временные ряды не стационарны. Это также соответствует классической гипотезе о «блуждающих» ценах на финансовых рынках.
